```python
import pandas as pd
import seaborn as sns; sns.set();
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
%pwd
```

```python
#opening the data set and storing in Pandas dataframe "df"
f = open('/home/localadmin/business.csv','r')
df = pd.read_csv(f);
counties = df.iloc[:,1].astype('category')
uniques = pd.unique(counties)
```

```python
#using PCA dimenstionality reduction on Captial investments, new jobs, revenue to generalize "success"
def calculate(s,writer):
    pca = PCA(n_components=1)
    pca.fit(df.iloc[:,6:9])
    tmp = df[df['County']== s]
    impact_pca = pd.DataFrame(pca.transform(tmp.iloc[:,6:9].values))
    codes = pd.DataFrame(tmp.iloc[:,12].values)
    df_new = pd.concat([codes,impact_pca],axis=1)
    df_new.columns = ["NAICS Code","Success"]
    ret = df_new.groupby("NAICS Code")["Success"].mean().reset_index().sort_values("Success",ascending=False)
    ret.to_excel(writer,sheet_name=s)
    return ret
```

```python
with pd.ExcelWriter('output.xlsx') as writer:
    for name in uniques:
        print(calculate(name,writer))
        


```
